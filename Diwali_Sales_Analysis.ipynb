```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv("Diwali Sales Data.csv", encoding='unicode_escape') 
#to avoid an encoding error use unicode_escape 
```

```python
df
```

# 1.Data Cleaning

```python
df.info()
```

# Delete unrelated/Blanked Columns

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

# check the null values

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
pd.isnull(df).sum()
```

```python
#change the datatype of Amount column from float to int
df['Amount']=df['Amount'].astype('int')
```

# 2.Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:  #label over the bar
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Gender",y="Amount", data=sales_gen,hue='Gender',palette='Set2')
plt.show()
```

*From above graph we can conclude that most of the buyers are females and even the purchasing power of females are greater*

# Age

```python
df.columns
```

```python
ax=sns.countplot(x="Age Group",data=df,hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sales_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age,hue='Age Group',palette='prism_r')
plt.show()
```

*From above graph we can see that most of the buyer are of the age group between 26-35 years females*

# State

```python
df.columns
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x="State",y="Orders",data=sales_state,hue='State',palette='prism')
plt.show() 
```

```python
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x="State",y="Amount",data=sales_state,hue='State',palette='rocket')
plt.show() 
```

*From above graph we can conclude that most of the orders and sale/amounts are from Uttar Pardesh, Maharashtra and Karnatka respectively*

# Marital Status

```python
df.columns
```

```python
ax=sns.countplot(x='Marital_Status',data=df,hue='Marital_Status',palette='prism')
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_mar=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
x=sns.barplot(x='Marital_Status',y='Amount',data=sales_mar,hue='Gender',palette='rainbow')
for bars in x.containers:
    x.bar_label(bars)
plt.show()
```

*From above we can see the that most of the buyers are the married (Women) and they have high purchasing power*

# Occupation

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df, hue='Occupation',palette='magma')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sales_occ,hue='Occupation',palette='inferno')
```

From above graph we can see that most of the buyers are working in IT,Health Care and Aviation Sector

# Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Product_Category',data=df, hue='Product_Category',palette='magma')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sales_occ=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sales_occ,hue='Product_Category',palette='nipy_spectral_r')
```

*From above graphwe can see that most of the sold products are from Food, Clothing and Electronics category*
